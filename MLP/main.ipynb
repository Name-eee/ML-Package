```python
# import
import numpy as np
from tqdm import tqdm
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.datasets import load_iris, load_breast_cancer, load_digits
from config import *
from model import *
```

```python
# load dataset from sklearn && check
dataset = load_breast_cancer()
# train_X = datasets.MNIST(root="../mnist", train=True, download=True)
# test_X = datasets.MNIST(root="../mnist", train=False, download=True)
print("keys:", dataset.keys())
print("len:", len(dataset.data))
print("data:", dataset.data[0].shape)
print("target:", dataset.target[:5])
print("data range:", np.max(dataset.data[0]), np.min(dataset.data[0]))
print("target range:", (np.min(dataset.target), np.max(dataset.target)))
print("target names:", dataset.target_names)
print("feature_num:", len(dataset.feature_names))
```

```python
# config
epoch=500
train_data, test_data=train_test_split(dataset)
data, target=dataloader(train_data, batch_size=32)
model=MLP(lr=0.05, initr=0.05, feature_num=30, variety=2)
# model=MLP(lr=0.05, initr=0.5)
```

```python
# train
loss_log=[]
acc_log=[]
epoch_log=list(range(epoch))
for i in tqdm(range(epoch)):
    for x,y in zip(data, target):
        y_pred, loss=model.train(x, y)
    # print(f"{np.max([np.max(i) for i in model.grad_log]):.4f}")
    # print(f"{np.min([np.min(i) for i in model.grad_log]):.4f}")
    loss_log.append(loss)
    acc_log.append(cal_accuracy(y_pred, y))
    # if i%20==19:
    #     print(f"Epoch {i}: loss:{loss}, accuracy:{cal_accuracy(y_pred, y)}")
print()
print(f"Final: loss:{loss}, accuracy:{cal_accuracy(y_pred, y)}")
# print(y_pred[:5])
draw_logs(epoch_log, loss_log, acc_log, title='iris classify')
```

```python
# test
x_test, y_test=test_data['data'], test_data['target']
# print(x_test[:5])
# print(y_test[:5])
y_pred, loss=model.test(x_test, y_test)
print(f"Test: loss:{loss}, accuracy:{cal_accuracy(y_pred, y_test)}")
y_pred = np.argmax(y_pred, axis=1)
print(y_pred[:5])
```

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

```python
print(classification_report(y_test, y_pred))
```
